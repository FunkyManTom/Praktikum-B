```julia
using PyPlot
using PyCall

PyCall.pyimport_conda("mpmath", "mpmath")

P(n,t)=t.^n./(factorial(n)).*exp.(-t)
```

```julia
t=collect(0:0.01:10)

rc("text", usetex=true)
rc("font", family="serif")

plot(t,P(0,t), c="black", label="n=0")
plot(t,P(1,t), c="crimson", label="n=1")
plot(t,P(2,t), c="black", ls="--", label="n=2")
plot(t,P(3,t), c="crimson", ls="--", label="n=3")

legend()
xlabel(L"$t$")
ylabel(L"$P_n$")

#savefig("Intervallverteilung.pdf", bbox_inches="tight")
```

```julia
#POISSON VERTEILUNG

poi1=[]

for line in eachline("B_Q67_500.txt")
    push!(poi1,parse(Int32,line))
end

##############
1=0

for i in 1:length(poi1)
    1=1 + poi1[i]*i
end
1=1/sum(poi1)

1=0

for j in 1:length(poi1)
    1=1 + poi1[j]*(j-1)^2
end
1=1/sum(poi1)

##########################
n=collect(range(1, stop=21))
P2(n,)=sum(poi1)*1^n*exp(-1) / (factorial(big(n)))

plotpoi=[]
plotn=[]
n2=[]

for i in 1:length(poi1)
    push!(plotpoi,poi1[i])
    push!(plotpoi,poi1[i])
    
    push!(plotn,i-0.5)
    push!(plotn,i+0.5)
    
    push!(n2,i)
    push!(n2,i)
end
   


plot(plotn,plotpoi, label="Experimentell", c="crimson")
plot(plotn,P2.(n2,1), label="Theoretisch", c="black")
xlabel(L"Zerflle pro $0.08s$")
ylabel("Zhlungen")
legend()

#savefig("Q67_500V_Poi.pdf", bbox_inches="tight")

println("N=", sum(poi1))
println("=", 1)
println("=", 1)
```

```julia
#GAUSSVERTEILUNG#

#Experimentell

gau1=[]

for line in eachline("G_500_5.txt")
    push!(gau1,parse(Int32,line))
end

plotgau=[]
plotn=[]
n2=collect(range(200, stop=330, step=0.1))

for i in 1:length(gau1)
    push!(plotgau,gau1[i])
    push!(plotgau,gau1[i])
    
    push!(plotn,i-0.5)
    push!(plotn,i+0.5)
    
end

for i in 1:25
    push!(plotgau,0)
    push!(plotgau,0)
    
    push!(plotn,plotn[end])
    push!(plotn,plotn[end]+1)  
    
end

for i in 1:200
    popfirst!(plotgau)
    popfirst!(plotgau)
    
    popfirst!(plotn)
    popfirst!(plotn)  
       
end

#Theoretisch

m=0

for i in 1:length(gau1)
    m=m + gau1[i]*i
end
m=m/sum(gau1)
N=sum(gau1)


PGau(x,m)=1/sqrt(2*pi*m)*sqrt(N^2+m^2)*exp(-(x-m)^2/(2*m))


plot(plotn,plotgau, label="Experimentell", c="crimson")
plot(n2,PGau.(n2,m),label="Theoretisch", c="black")
xlabel(L"Zerflle pro $5s$")
ylabel("Zhlungen")
legend()

println("N=", sum(gau1))
println("m=", m)

savefig("500V_Gauss.pdf", bbox_inches="tight")
```
